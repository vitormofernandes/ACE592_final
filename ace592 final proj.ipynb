```python
# Define the search term and the date_since date as variables
search_words = "$AAPL"
date_since = "2021-04-20" #YYYY-MM-DD
end_date = "2021-04-21"

# Collect tweets
tweets = tw.Cursor(api.search,
                       q=search_words,
                       lang="en",
                       since=date_since,
                        until=end_date).items(50000)


twts_we_want = [[tweet.user.screen_name, tweet.user.location, tweet.text, tweet.created_at] for tweet in tweets]

#Pandas df with user and location
AAPL = pd.DataFrame(data=twts_we_want, 
                    columns=['user', "location", "text", "create_at"])
AAPL.head(5)
```
